# **Calculate Cramer's V matrix for cat cols:**

```python
from scipy.stats import chi2_contingency
# Function to check for zero variance in categorical variable
def has_zero_variance(series):
    return len(series.unique()) == 1

# Function to calculate Cramer's V with handling for zero variance=unique
def cramers_v(x, y):
    if has_zero_variance(x) or has_zero_variance(y):
        return np.nan  # Return NaN if either variable has zero variance
    else:
        confusion_matrix = pd.crosstab(x, y)
        chi2, _, _, _ = chi2_contingency(confusion_matrix)
        n = confusion_matrix.sum().sum()
        phi2 = chi2 / n
        r, k = confusion_matrix.shape
        phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
        rcorr = r - ((r - 1) ** 2) / (n - 1)
        kcorr = k - ((k - 1) ** 2) / (n - 1)
        return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# Calculate Cramer's V matrix for all pairs of categorical variables
cat_cols = df.select_dtypes(include=['object']).columns
cramers_v_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols, dtype=float)
for i in range(len(cat_cols)):
    for j in range(i, len(cat_cols)):
        if i != j:
            v = cramers_v(df[cat_cols[i]], df[cat_cols[j]])
            cramers_v_matrix.at[cat_cols[i], cat_cols[j]] = v
            cramers_v_matrix.at[cat_cols[j], cat_cols[i]] = v

# Display Cramer's V matrix
print("Cramer's V Matrix:")
print(cramers_v_matrix)
```

```python
# Identify highly associated columns based on a threshold (e.g., 0.5)
threshold = 0.5
highly_associated_cols = set()
for i in range(len(cat_cols)):
    for j in range(i + 1, len(cat_cols)):
        if cramers_v_matrix.iloc[i, j] >= threshold:
            highly_associated_cols.add(cat_cols[j])

# Drop highly associated columns from the DataFrame
df.drop(columns=highly_associated_cols, inplace=True)

# columns to drop
print(f"Highly associated columns dropped: {highly_associated_cols}")

# Display the remaining columns after dropping highly associated columns
print(df.select_dtypes(include=['object']).columns.tolist())
```

# **Featuring Engineering:**

# **Reducing the uniqe values in emp_title col:**
annual_inc col is numerical so dont scale it before this step

```python
# Define thresholds based on overall median income
overall_median_income = df['annual_inc'].median()
low_income_threshold = overall_median_income / 2
high_income_threshold = overall_median_income * 1.5


# Create a function to categorize jobs based on income
def categorize_job(title):
    annual_income = df[df['emp_title'] == title]['annual_inc'].mean()# mean of annual_inc after filtering according to the given job title
    if annual_income < low_income_threshold:
        return 'low_income_job'
    elif annual_income < high_income_threshold:
        return 'moderate_income_job'
    else:
        return 'high_income_job'

# Apply the categorization to the 'emp_title' column
df['emp_title'] = df['emp_title'].apply(categorize_job)
```

# **Hyperparameter Tuning:**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.datasets import make_classification
from scipy.stats import randint
import numpy as np

# Define RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Define parameter distributions
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
    'max_features': ['auto', 'sqrt', 'log2']
}

# Define scoring metrics
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='weighted'),
    'recall': make_scorer(recall_score, average='weighted'),
    'f1_score': make_scorer(f1_score, average='weighted')
}

# Perform Randomized Search Cross-Validation
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_dist, scoring=scoring, refit='f1_score',
                                   cv=5, verbose=2, n_iter=100, random_state=42, n_jobs=-1)
random_search.fit(X_train, y_train)

# Get all results including mean and std of each score metric
results = random_search.cv_results_

# Print out accuracy, precision, recall, and F1 score for each candidate hyperparameter set
for mean_acc, mean_prec, mean_rec, mean_f1, params in zip(results['mean_test_accuracy'],
                                                          results['mean_test_precision'],
                                                          results['mean_test_recall'],
                                                          results['mean_test_f1_score'],
                                                          results['params']):
    print(f"Accuracy: {mean_acc:.4f}, Precision: {mean_prec:.4f}, Recall: {mean_rec:.4f}, F1 Score: {mean_f1:.4f} - {params}")

# Print best parameters and best score
print("\nBest parameters found:", random_search.best_params_)
print("Best F1 score found:", random_search.best_score_)
```

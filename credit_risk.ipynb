# **Imputing the NAN values:**

```python
from sklearn.impute import SimpleImputer

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
numerical_cols = numerical_cols.drop('mths_since_last_delinq')

categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# Impute low missing value columns for numerical data
num_imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = num_imputer.fit_transform(df[numerical_cols])

# Impute low missing value columns for categorical data
cat_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])
```

# **Recuding the unique values in grade col:**

```python
# Assuming df is your DataFrame
grade_map = {
    'A': 'A',
    'B': 'A',
    'C': 'B',
    'D': 'B',
    'E': 'C',
    'F': 'C',
    'G': 'C'
    }

df['grade'] = df['grade'].map(grade_map)
```

# **Reducing the uniqe values in emp_title col:**

```python
# Define thresholds based on overall median income
overall_median_income = df['annual_inc'].median()
low_income_threshold = overall_median_income / 2
high_income_threshold = overall_median_income * 1.5


# Create a function to categorize jobs based on income
def categorize_job(title):
    annual_income = df[df['emp_title'] == title]['annual_inc'].mean()# mean of annual_inc after filtering according to the given job title
    if annual_income < low_income_threshold:
        return 'low_income_job'
    elif annual_income < high_income_threshold:
        return 'moderate_income_job'
    else:
        return 'high_income_job'

# Apply the categorization to the 'emp_title' column
df['emp_title'] = df['emp_title'].apply(categorize_job)
```

# **Reducing the uniqe values in emp_length col:**

```python
# Create a function to categorize employment length
def categorize_emp_length(length):
    if length == '< 1 year' or length == '1 year':
        return 'short'
    elif length in ['2 years', '3 years', '4 years', '5 years']:
        return 'medium'
    else:
        return 'long'

# Apply the categorization to the 'emp_length' column
df['emp_length'] = df['emp_length'].apply(categorize_emp_length)
```
